```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.lamudi.com.ph/apartment/rent/'

page = requests.get(url)


soup = BeautifulSoup(page.content,'html.parser')
soup
```

```python
title_list=[]
location_list = []
price_list=[]
description_list=[]
def append_title(title):
    for t in title:
        title_list.append(t.get_text().strip())
    
def append_location(locations):
    for loc in locations:
        location_list.append(loc.get_text().strip())
    
def append_price(prices):
    for price in prices:
        price_list.append(price.get_text().strip())
        
def append_description(description):
    for desc in description:
        description_list.append(desc.get_text().strip())

```

```python

for i in range(50):
    
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    title = soup.find_all('h2',{'class':'ListingCell-KeyInfo-title'})
    locations = soup.find_all('span',{'class':'ListingCell-KeyInfo-address-text'})
    prices = soup.find_all('span',{'class':'PriceSection-FirstPrice'})
    description = soup.find_all('div',{'class':'ListingCell-shortDescription'})
    append_title(title)
    append_location(locations)
    append_price(prices)
    append_description(description)
    url = 'https://www.lamudi.com.ph/apartment/rent/?page='+ str(i)
```

```python
dataset =pd.DataFrame(list(zip(title_list, location_list,price_list,description_list)),columns=['Title','Location','Price','Description'])
dataset
```

```python
dataset.to_excel('apartment_scraper.xlsx',index=False)
```
